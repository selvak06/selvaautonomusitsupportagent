```python
def factorial(n):
  if(n==1):
    return 1
  else :
    return n * factorial(n-1)

factorial(5)
```

```python
from abc import ABC, abstractmethod
class Animal:
  def __init__(self,name) :
    self.name = name

  @abstractmethod
  def move(self):
    pass

class Human(Animal) :
  def mopve(self) :
    print (f"{self.name} walks on two legs")

class Fish(Animal) :
  def mopve(self) :
    print (f"{self.name} swims in water")

john= Human("john")
nemo = Fish("nemo")

john.move()
nemo.move()
```
